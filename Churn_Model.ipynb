### Importing Libraries

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense
```

### Importing Data

```python
data = pd.read_csv("data.csv")
```

```python
data
```

```python
data.info()
```

```python
data.describe().T
```

### Drop unique columns from data

```python
data = data.iloc[:,3:]
```

```python
data
```

### *Geography* column One Hot Encoding

```python
ohe = OneHotEncoder()
ohe_data = ohe.fit_transform(data.iloc[:,[1]]).toarray()
ohe_data
# France = 0
# Germany = 1
# Spain = 2
```

```python
data["isFrance"] = ohe_data[:,[0]]
data["isGermany"] = ohe_data[:,[1]]
data["isSpain"] = ohe_data[:,[2]]
data.drop("Geography",axis=1, inplace=True)
```

```python
data
```

###  Gender column Laber Encoding

```python
lb = LabelEncoder()
lb_gender = lb.fit_transform(data[["Gender"]])
data["Male"] = lb_gender
data.drop("Gender", axis=1, inplace=True)
```

```python
data
```

### Standardization

```python
standardization_columns = ["CreditScore", "Age", "Tenure", "Balance", "EstimatedSalary"]
nan_standardization_columns_values = data.drop(standardization_columns, axis=1)
standardization_columns_values = data[standardization_columns]
standardization_columns_values
```

```python
nan_standardization_columns_values
```

```python
stds = StandardScaler()
std_values = stds.fit_transform(standardization_columns_values)
std_values_df = pd.DataFrame(std_values, columns=standardization_columns)
std_values_df
```

```python
scale_data = pd.concat([std_values_df, nan_standardization_columns_values],axis=1)
scale_data
```

### Train Test Split

```python
Y = data[["Exited"]]
X = data.drop("Exited", axis=1)
```

```python
trainX, testX, trainY, testY = train_test_split(X,Y,test_size=0.33, random_state=42)
```

```python
model = Sequential()
model.add(Dense(24, activation="sigmoid", input_shape=(12,)))
model.add(Dense(12, activation="sigmoid"))
model.add(Dense(1, activation="sigmoid"))
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
model.fit(trainX,trainY,epochs=50)
```

```python
predY = model.predict(testX)
```

```python
from sklearn.metrics import accuracy_score
```

```python
predY[predY>0.5] = 1
predY[predY<0.5] = 0
predY[predY==1]
```

```python
accuracy_score(predY, testY)
```
